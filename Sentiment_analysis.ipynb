```python
import pandas as pd
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Load data from CSV file
data = pd.read_csv('/content/drive/MyDrive/Sentiment analysis /output.csv')

# Initialize SentimentIntensityAnalyzer
sid_obj = SentimentIntensityAnalyzer()

# Function to calculate sentiment scores
def sentiment_scores(sentence):
    # Polarity_scores method of SentimentIntensityAnalyzer object gives a sentiment dictionary
    sentiment_dict = sid_obj.polarity_scores(sentence)

    # Return sentiment scores
    return sentiment_dict['neg'], sentiment_dict['neu'], sentiment_dict['pos'], sentiment_dict['compound']

# Lists to store sentiment scores
negative_scores = []
neutral_scores = []
positive_scores = []
sentiments = []

# Process each caption from the CSV file
for caption in data['comments']:
    # Preprocess caption text
    caption = str(caption)
    caption = caption.replace("\\n", " ")
    cleaned_caption = re.sub(r'[^a-zA-Z0-9 ]', r'', caption)

    # Perform sentiment analysis
    neg, neu, pos, compound = sentiment_scores(cleaned_caption)

    # Append sentiment scores to lists
    negative_scores.append(neg * 100)
    neutral_scores.append(neu * 100)
    positive_scores.append(pos * 100)

    # Determine overall sentiment
    if compound >= 0.05:
        sentiments.append("Positive")
    elif compound <= -0.05:
        sentiments.append("Negative")
    else:
        sentiments.append("Neutral")

# Add new columns to the DataFrame
data['negative'] = negative_scores
data['neutral'] = neutral_scores
data['positive'] = positive_scores
data['sentiment'] = sentiments

# Save DataFrame back to CSV
data.to_csv('/content/drive/MyDrive/Sentiment analysis /output_with_sentiments.csv', index=False)
```
